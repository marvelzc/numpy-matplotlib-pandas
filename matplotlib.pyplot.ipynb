### 设置坐标轴与轴标签

```python
x = np.linspace(-1, 1, 50)
y1 = x
y2 = x**2

plt.plot(x, y1)
plt.plot(x, y2)

#设置坐标轴长度
plt.xlim((-1, 2))
plt.ylim((-1, 2))
# 设置坐标轴标签
plt.xlabel('I am x')
plt.ylabel('I am y')
# 设置坐标轴刻度
ticks = np.linspace(-1, 2, 5)
plt.xticks(ticks)
# plt.yticks([-1, -0.25, 0.5, 1.25, 2.],['not good','normal','good','better','best'])
plt.yticks([-1, -0.25, 0.0, 1.25, 2.],[r'$not\ good$',r'$normal$',r'$good$',r'$better$',r'$best$'])
```

### 设置图形边框axis

```python
x = np.linspace(-1, 1, 50)
y1 = x
y2 = x**2

plt.plot(x, y1)
plt.plot(x, y2)

plt.xlim((-1, 2))
plt.ylim((-1, 2))

plt.xlabel('I am x')
plt.ylabel('I am y')

ticks = np.linspace(-1, 2, 5)
plt.xticks(ticks)
plt.yticks(ticks)

# gca = 'get current axis'
ax = plt.gca()
ax.spines['left'].set_color('r')
ax.spines['right'].set_color('y')
ax.spines['top'].set_color('b')
ax.spines['bottom'].set_color('g')

# 移动坐标轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
```

### legend

```python
x = np.linspace(-1, 1, 50)
y1 = x
y2 = x**2

plt.xlim((-1, 2))
plt.ylim((-1, 2))

ticks = np.linspace(-1, 2, 5)
plt.xticks(ticks)
plt.yticks(ticks)


# plt.plot(x, y1, label='x')
# plt.plot(x, y2, label='x**2')
# plt.legend()

l1,=plt.plot(x, y1, label='x')
l2,=plt.plot(x, y2, label='x**2')
plt.legend(handles=[l1,l2], labels=['aaa','bbb'], loc='best')
```

### scatter

```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib tk
X = np.random.normal(0, 1, 1024)
Y = np.random.normal(0, 1, 1024)
T = np.arctan2(Y, X)    # for color later on

# plt.scatter(X, Y, s=75, c=T, alpha=.5, edgecolors='none')
plt.scatter(X, Y, s=75, c=T, cmap=plt.cm.hsv, alpha=.5, edgecolors='none')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xticks(())
plt.yticks(())
```

### bar

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
for i in range(n):
    color=np.random.choice(['#99ffff','#ff9999','#99ff99','#ffff99','#ff99ff'])
    plt.bar(X[i], -Y1[i], facecolor=color, edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x, y, '%.2f' % y, ha='center', va='bottom')
    plt.text(x, -y, '%.2f' % y, ha='center', va='top')
```

### contourf

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)

# use plt.contourf to filling contours
# X, Y and value for (X,Y) point
plt.contourf(X, Y, f(X, Y), 10, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
plt.clabel(C, inline=True, fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```
