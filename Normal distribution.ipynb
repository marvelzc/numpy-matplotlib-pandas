```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a=np.zeros(1000)
for i in range(len(a)):
    for _ in range(1000):
        a[i] += np.random.choice([-1,1])

c={}
for i in set(a):
    c[i]=list(a).count(i)

plt.bar(c.keys(), c.values())
z = pd.Series(c)
plt.plot(z)
```
