```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib tk
```

```python
x=np.arange(10)
y=np.arange(10,0,-1)
```

### 简单绘图的三种方式

```python
plt.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

### 颜色、标记和线型

```python
plt.plot(x,y,color='k',linestyle='--',marker='s',linewidth=3)
```

```python
plt.plot(x,y, 'r--', x,y*2, 'bs', x,y*3, 'g^')
```

```python
plt.scatter(x,y, 200, 'r')
```

### 多视窗绘图

```python
fig=plt.figure()
ax1=fig.add_subplot(221)
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)
ax4.plot(x,y)
```

```python
#fig=plt.figure()
#ax1=fig.add_subplot(221)
fig,ax=plt.subplots(2,2)
ax[1,0].plot(x,y)
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
ax[1,0].scatter(x,y, 200)
plt.subplots_adjust(wspace=0, hspace=0)
```

### 标题、标签、图例

```python
fig,ax=plt.subplots()
ax.plot(x,y,'rs:',label='line')
ax.legend(loc='best') #设置图例项的位置
ax.set(title="My First Plot",xlabel="x",ylabel="y") #标题、坐标轴的标签
ax.grid() #添加网格线
```
